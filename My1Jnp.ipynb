```python
import pandas as pd
data_frame=pd.read_csv("data.csv")
data_frame.shape
```

```python
def calc(x):
    s = x+x
    return s

calc(5)

data_frame.head()
```

```python
import seaborn as sns
sns.set()

graph = sns.scatterplot(x='Marking', y='Age', data=data_frame)
graph
```

```python
from bokeh.plotting import figure, show
from bokeh.models import HoverTool

TOOLTIPS=HoverTool(tooltips=[
    ("index", "$index"),
    ("(Age,Marking)", "(@Age, @Marking)"),
    ("Name", "@Name")]
                  )

p= figure(title="simple ligne", x_axis_label='x', y_axis_label='y', plot_width=700, plot_height=700, tools=[TOOLTIPS])
p.circle('Age','Marking', size=10, source=data_frame)
show(p)
```
