```python
import pandas as pd
import numpy as np
import pyodbc
```

```python
df = pd.DataFrame({'CONTRAT':['C1', 'C1','C1','C2','C2'], 'NC':['1', '1','1','2','2'],
                   'COM':[10.0,11.0,12.0,21.0,22.0],'LIGQ':[110.0,111.0,121.0,211.0,221.0],
                   'GARANTIE':['G01', 'G02','G03','G01','G01'],
                   'HT':[10.0,11.0,12.0,21.0,22.0],'TTC':[15.0,16.0,17.0,29.0,30.0],
                   'INFO1':['info1', 'info1','info1','info101','info101'],'INFO2':['info21', 'info21','info21','info211','info211'],'INFO3':['a1', 'a1','a1','ab1','ab1'],
                   'INFO4':['info1', 'info1','info1','info101','info101'],'INFO5':['info21', 'info21','info21','info211','info211'],'INFO6':['a1', 'a1','a1','ab1','ab1'],
                   'INFO7':['info1', 'info1','info1','info101','info101'],'INFO8':['info21', 'info21','info21','info211','info211'],
                  });df
```

```python
df.columns
```

```python
df1_com = df.copy()
df1_com["sumcom"]=df.loc[(df["GARANTIE"]=="G01"),["COM"]].sum(axis=1);df1_com.fillna(0)
```

```python
df1_lig = df1_com.copy()
df1_ligq["sumligq"]=df.loc[(df["GARANTIE"]=="G01"),["LIGQ"]].sum(axis=1);df1_com.fillna(0)
```

```python
df1_ligq = df1_ligq.groupby(by=['CONTRAT','NC']).sum().groupby(level=[0]).cumsum();df1_ligq
```

```python
df1_ligq=df1_ligq.reset_index();df1_ligq
```

```python
df1_htgs = df.pivot_table(index=['CONTRAT', 'NC'],  aggfunc='sum'    ,                       
                             columns =['GARANTIE'], 
                             values =['HT']).fillna(0);df1_htgs
```

```python
df1_htgs=df1_htgs.reset_index()
```

```python
df1_htgs.columns
```

```python
df1_infos= df[['CONTRAT', 'NC','INFO1','INFO2','INFO3' ,'INFO4','INFO5','INFO6','INFO7','INFO8',]];df1_infos
```

```python
df1_infos=df1_infos.drop_duplicates().reset_index();df1_infos
```

```python
df2=pd.DataFrame(columns=["CONTRAT"])
df2["CONTRAT"] = df1_htgs[('CONTRAT',    '')]
df2["NC"] = df1_htgs[('NC',    '')]
df2["HTG01"] = df1_htgs[('HT',    'G01')]
df2["HTG02"] = df1_htgs[('HT',    'G02')]
df2["HTG03"] = df1_htgs[('HT',    'G03')];df2
```

```python
df3=df2.merge(df1_infos, on=['CONTRAT', 'NC']);df3
```
