```python
df_long_ttc = df.pivot_table(index=['CONTRAT','NC','COM'],
                             columns =['GARANTIE'], 
                             values =['TTC']).reset_index().fillna(0).rename(columns={'G01':'G01ttc','G02':'G02ttc','G03':'G03ttc','G04':'G04ttc',
                             'G05':'G05ttc','G06':'G06ttc','G07':'G07ttc','G08':'G08ttc','GFG':'GFGttc','SIN':'SINttc'});df_long_ttc
```

```python
df_long_ht = df.pivot_table(index=['CONTRAT','NC','COM'],
                             columns =['GARANTIE'], 
                             values =['HT']).reset_index().fillna(0).rename(columns={'G01':'G01ht','G02':'G02ht','G03':'G03ht','G04':'G04ht',
                             'G05':'G05ht','G06':'G06ht','G07':'G07ht','G08':'G08ht','GFG':'GFGht','SIN':'SINht'});df_long_ht
```
